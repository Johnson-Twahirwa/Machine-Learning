1. Create a list called Students with names and marks for 5 students

```python
Students = {'John': 78, 'Joe': 81, 'Json': 88, 'Jonah': 77, 'Joy': 86}
```

```python
Students
```

2. Extract the name and marks of the first 3 students

```python
first_three = list(Students.items())[:3]
for names, marks in first_three:
    print(names, marks)
```

3. Extract the name and marks of the last 2 students

```python
last_three = list(Students.items())[2:]
for names, marks in last_three:
    print(names, marks)
```

4. Extract the name of all the students and marks

```python
Students
```

5. Extract the last item in the list

```python
last_item = list(Students.items())[-1:]
for names, marks in last_item:
    print(names, marks)
```

6. Create a new list called list_2 with names and marks of 2 students

```python
list_2 = {'Kelly': 82, 'Nelly': 69}
```

7. Join the two lists and call it all_students

```python
all_students = {**Students, **list_2}
all_students
```

```python
all_students['John'] = 999
all_students['Nelly'] = 122
all_students
```
