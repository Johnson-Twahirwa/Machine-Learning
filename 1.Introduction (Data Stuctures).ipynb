2. Tuple: A simple immutable (unchangeable) ordered sequence of items
- Items can be of mixed types, including collection types
- Uses parentheses ()
- Items are accessed using their position in the tuple
- It is ordered

```python
tu = (23, 'abc', 4.56, (2,3), 'def')
```

```python
tu[2] # Item on index 2 
```

```python
tu[-1] # Last item on the list
```

```python
tu[::-1] # Reverse order
```

```python
type(tu) 
```


Slicing


```python
tu[1:4] # Slice from index 1 to 3
```

```python
tu[1:] # Slice from index 1 to last item
```


3. Dictionaries: Data structures which are used to store values.
    - Which are accessed by their keys.
    - Uses curly braces {}

```python
dict_1 = {'Ten': 10, 'Twenty': 20, "Thirty": 30}
```

```python
dict_1
```

```python
type(dict_1)
```

```python
dict_1['Thirty']
```


Other functons:
* dict.clear() - Removes all the elements from dictionary dict
* dict.copy() - Returns a copy of dictionary dict
* dict.items() - Returns a list containing a tuple for each key-value pair in dictionary dict
* dict.get(k) - Returns the value of the specified key k from dictionary dict
* dict.keys() - Returns a list containing all the keys of dictionary dict
* dic.pop(k) - Removes the element with the specified key k from dictionary dict
* dic.values() - Returns a list of all the values in dictionary dict


```python
dict_2.keys()
```

```python
dict_2.values()
```

```python
dict_2.copy()
```


Convert two lists into a Dictionary


```python
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
```

```python
combined_dict = dict(zip(keys,values))
```

```python
combined_dict
```


Merge two Python dictionaries


```python
dict_1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30}
dict_2 = {'Forty': 40, 'Fifty': 50, 'Sixty': 60}
```

```python
merged_dict = dict(**dict_1, **dict_2)
merged_dict
```

```python
type(merged_dict)
```
